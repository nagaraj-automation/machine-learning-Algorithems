# **Chi Squre test** applied when there two catogorical variables from single population.It is used to determine an association between two **variables**

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("sample_data/tips.csv")
```

```python
df.info
```

```python
df.describe()
```

```python
import scipy.stats as stats
```

```python
df.head()
```

```python
df_table=pd.crosstab(df['sex'],df['smoker'])
df_table
```

```python
df_table.values
```

```python
observed_values=df_table.values
print("observed_values \n",observed_values)
```

```python
val=stats.chi2_contingency(df_table)
val
```

```python
expected_values=val[3]
expected_values
```

```python
no_of_rows=len(df_table.iloc[0:2,0])
no_of_rows
```

```python
no_of_cols=len(df_table.iloc[0,0:2])
no_of_cols
```

```python
#degree of freedom
dof=(no_of_rows-1)*(no_of_cols-1)
dof
```

Lets assume alpha value is 0.05(significance value) and using the **chisquare** formula as below!




chisqaure_formula.png

```python
alpha=0.05
```

```python
chi_square=sum([(o-e)**2./e for o,e in zip(observed_values,expected_values)])
chi_square_stat=chi_square[0]+chi_square[1]
chi_square_stat
```

```python
from scipy.stats import chi2
critical_value=chi2.ppf(q=1-alpha,df=dof)
critical_value
```

lets find the p value

```python
p_value=1-chi2.cdf(x=chi_square_stat,df=dof)
print("alpha value",alpha)
print("p_value", p_value)
print("critical value",critical_value)
```

```python
if chi_square_stat>=critical_value:
  print("Reject H0 There is a relation between two categorical value")
else:
  print("Retain H0 There is no relation between two categorical value")
if p_value<=alpha:
  print("Reject H0 There is a relation between two categorical value")
else:
  print("Retain H0 There is no relation between two categorical value")
```
